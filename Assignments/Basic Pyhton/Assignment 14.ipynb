# 3. What is a box tuple, and how does it work?

```python
# Ans : A box tuple is a tuple value of four integers: the left-edge x-coordinate, the top-edge y-coordinate, 
#       the width, and the height, respectively.
```

# 5. What method would you call to get Image object for a 100×100 image, excluding the lower-left quarter of it?

```python
# Ans: imageObj.crop((0, 50, 50, 50)).
#      Notice that you are passing a box tuple to crop(), not four separate integer arguments.
from PIL import Image
im = Image.open('okayokay.jpeg')
imageObj = im.crop((0, 50, 50, 50))
print(imageObj)
```
